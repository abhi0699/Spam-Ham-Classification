```python
import numpy as np
import pandas as pd
import os
```

```python
Df = []
```

```python
# Reading the files
import os
base_path = "D:/Abhishek/CSE/bare/"
all_file = os.listdir(base_path)
for parts in all_file:
    files = base_path + parts
    all_txt = os.listdir(files)
    
    for txt in all_txt:
        txt_dir = files + "/" + txt
        with open(txt_dir, 'r') as txt_file:
            j = txt_file.read()
            Df.append(j)
```

```python
# Converting List to a string named new
# tokens = Vocabulary#1
new = " "
for x in Df:
    new+=x

# Tokenizing the string
import nltk
nltk.download('punkt')
tokens = nltk.word_tokenize(new)
nltk.download('stopwords')
```

```python
# Removal Of Stopwords
# tokens = Vocabulary#2
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
tokens = [w for w in tokens if not w in stop_words]

  
```

```python
# Lemmatization of Words
# new_1 = Vocabulary#3
from nltk.stem import WordNetLemmatizer 
nltk.download('wordnet')
  
lemmatizer = WordNetLemmatizer() 
for i in range(len(tokens)):
    tokens[i]=lemmatizer.lemmatize(tokens[i])
```

```python
new_1 = ""
for x in tokens:
    new_1+=x
    new_1+=' '
    
```

```python
# Counting the no of words
from collections import Counter
cnt = dict(Counter(new_1.split()))
```

```python
len(cnt)
```

```python
# Setting low and high frequency to filter the vocabulary
# Vocabulary#4
new_2 = []
low = 2
high = 10000
for i in cnt:
    if cnt[i] < low or cnt[i]> high:
        new_2.append(i)
        
for i in new_2:
    del cnt[i]
```
